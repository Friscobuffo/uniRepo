```python
from time import time as ltime0316
from random import random as rnd0316

class ProgressBar():
    __fill = '█'

    def __init__(self, iterations, barLength = 40):
        self.__iterations = iterations
        self.__counter = 0
        self.__start = ltime0316()
        self.__end = False
        self.__freq10 = 1
        self.__bar_length = barLength-1
    
    def __fill_fun(self, val):
        if val==0.0: return '█'
        if val<0.125: return '▏'
        if val<0.25: return '▎'
        if val<0.375: return '▍'
        if val<0.5: return '▌'
        if val<0.625: return '▋'
        if val<0.75: return '▊'
        if val<0.875: return '▉'
        return '█'

    def __update(self):
        this_time = ltime0316()
        totDone = self.__counter
        totTime = this_time - self.__start
        freq = self.__freq(totDone / totTime)
        self.__freq10 = freq // 10
        if self.__freq10 == 0: self.__freq10 = 1
        self.__print_bar(freq, totTime)

    def next(self):
        self.__counter += 1
        if not self.__end:
            if self.__counter == self.__iterations:
                self.__end = True
                self.__update()
                return
            elif self.__counter <= 20:
                self.__update()
            else:
                if self.__counter % self.__freq10 == 0:
                    self.__update()

    def reset(self, iterations, barLength = 40):
        self.__init__(iterations, barLength)
    
    def __freq(self, freq):
        if freq > 10: return int(freq)
        return round(freq, 2)

    def __eta(self, freq):
        eta = int((self.__iterations-self.__counter) / freq)
        mins = ""
        if eta >= 60:
            mins = int(eta/60)
            mins = str(mins) + "mins "
            eta = eta%60
        return mins + str(eta)
    
    def __last_output(self, bar, totTime):
        if totTime >= 60:
            mins = int(totTime/60)
            secs = int(totTime%60)
            totalTime = str(mins) +" mins " + str(secs) + " seconds"
        else:
            totalTime = str(int(totTime)) + " seconds"
        return f'\r{bar} [100.00]% (total time: {totalTime})'+(20)*" "

    def __print_bar(self, freq, totTime):
        eta = self.__eta(freq)
        if type(freq) == int:
            frequency = str(freq)
        else:
            frequency = "{:.2f}".format(freq)
        percent = 100*(self.__counter / float(self.__iterations))
        spercent = "{:.2f}".format(percent)
        filledLength = self.__bar_length * self.__counter // self.__iterations
        factor = (self.__bar_length * self.__counter / self.__iterations) - (self.__bar_length * self.__counter // self.__iterations)
        # bar = '|' + (self.__fill * filledLength) + '-' * (self.__bar_length - filledLength) + '|'
        bar = '|' + (self.__fill * (filledLength)) + self.__fill_fun(factor) + ' ' * (self.__bar_length - filledLength) + '|'
        if not self.__end:
            print(f'\r{bar} [{spercent}]% (eta: {eta}s, frequency: {frequency}/s)'+7*" ", end = '', flush = True)
        else:
            print(self.__last_output(bar, totTime))
```

```python
from time import time, sleep
inizio = time()

tot = 5678
bar = ProgressBar(tot)
i=0
while i<tot:
    bar.next()
    if i<tot/2:
        sleep(0.001)
    else:
        sleep(0.01)
    
    i += 1
print(time()-inizio)
bar.reset(tot)
i=0
while i<tot:
    if i<tot/2:
        sleep(0.001)
    else:
        sleep(0.002)
    bar.next()
    i += 1
```
