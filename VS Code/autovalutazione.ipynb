```python
#domanda 4
import random
import pandas as pd
import numpy as np
from progress.bar import Bar

def lancioDadi():
    return random.randint(1,6) + random.randint(1,6)

def probabilita(v):
    if v<=7: return 100*(v-1)/36
    return 100*(13-v)/36

def simulaLanci(lanci):
    i = 0
    risultati = np.zeros(11, dtype=int)
    with Bar(max=lanci) as bar:
        for i in range(lanci):
            bar.next()
            risultati[lancioDadi()-2]+=1
    output = pd.DataFrame()
    for i in range(2,13):
        temp = pd.DataFrame({'simulata': round(100*risultati[i-2]/lanci,2),
                             'prevista': round(probabilita(i),2)},
                             index = [i])
        output = pd.concat([output,temp])
    return output

lanci = 1000000
result = simulaLanci(lanci)
result
```
